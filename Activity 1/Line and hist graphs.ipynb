## Line Graph

```python
import matplotlib.pyplot as plt
plt.plot(df["TransactionID"],df["txn_total"])
```

### Descriptive line Graph

```python
plt.plot(df["TransactionID"],df["txn_total"])
plt.xlabel("TransactionID")
plt.ylabel("txn_total")
plt.title("TransactionID vs total Transaction")
```

### Using Markers

```python
### using the marker
plt.plot(df["TransactionID"],df["txn_total"],"r--p")
plt.xlabel("TransactionID")
plt.ylabel("txn_total")
plt.title("TransactionID vs total Transaction")
```

### Including legends

```python
plt.plot(df["TransactionID"],df["txn_total"],"r--p",label="Transaction Amounts")
plt.xlabel("TransactionID")
plt.ylabel("txn_total")
plt.title("TransactionID vs total Transaction")
plt.legend()
```

### Rotating and padding labels alnog the axis

```python
plt.plot(df["TransactionID"],df["txn_total"],"r--p",label="Transaction Amounts")
plt.xlabel("TransactionID")
plt.ylabel("txn_total",rotation=0,labelpad=40)
plt.title("TransactionID vs total Transaction")
plt.legend()
```

### Adjusting the size of graph

```python
plt.figure(figsize=(15,7))
plt.plot(df["TransactionID"],df["txn_total"],"r--o",label="Transaction Amounts")
plt.xlabel("TransactionID")
plt.ylabel("txn_total",rotation=0,labelpad=40)
plt.title("TransactionID vs total Transaction")
plt.legend()
```

### Adjusting the position of the legend

```python
plt.figure(figsize=(15,5))
plt.plot(df["TransactionID"],df["txn_total"],"r--o",label="Transaction Amounts")
plt.xlabel("TransactionID")
plt.ylabel("txn_total",rotation=0,labelpad=40)
plt.title("TransactionID vs total Transaction")
plt.legend(loc="lower right")
```

### Adding marker values

#### 1. Using enumerate

```python
plt.figure(figsize=(15,5))
plt.plot(df["TransactionID"],df["txn_total"],"r--o",label="Transaction Amounts")
for index,value in enumerate(df["txn_total"]):
    plt.annotate(text=value,xy=(df["TransactionID"][index],df["txn_total"][index]),xytext=(8.0, -10.0),textcoords='offset points')
plt.xlabel("TransactionID")
plt.ylabel("txn_total",rotation=0,labelpad=40)
plt.title("TransactionID vs total Transaction")
plt.legend(loc="lower right")
```

#### 2.  Using zip

```python
plt.figure(figsize=(15,5))
plt.plot(df["TransactionID"],df["txn_total"],"r--o",label="Transaction Amounts")
for id,totalTrans in zip(df["TransactionID"],df["txn_total"]):
    plt.annotate(text=totalTrans,xy=(id,totalTrans),xytext=(8.0, -10.0),textcoords='offset points')
plt.xlabel("TransactionID")
plt.ylabel("txn_total",rotation=0,labelpad=40)
plt.title("TransactionID vs total Transaction")
plt.legend(loc="lower right")
```
